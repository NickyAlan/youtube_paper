```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = mX + b + noise
n = 100
X = np.random.rand(n,1)
y = 2.3*X + 5 + np.random.randn(n,1) 

plt.scatter(X, y, s=5, color='r')
plt.title(f'{n} point of data')
plt.xlabel('x')
plt.ylabel('y');
```

```python
# gradient descent 
# y = wx + b
def linear_gradient_descent(a=0.01,epochs=10):
    w, b = np.random.randn(2) # initial weight and bias
    for epoch in range(epochs) :
        y_pred = w*X + b 
        dw = (1/n)*np.sum((y_pred-y)*X) 
        db = (1/n)*np.sum(y_pred-y)
        b = b - (a*db) # update weight and bias
        w = w - (a*dw)
        mse = np.mean(np.square((w*X + b) - y)) # mse after update weight and bias
        if (epoch==0) or (epoch%100 == 0) or (epoch==(epochs-1)) :
            print(f'epoch {epoch} | w = {w:.6f} , b = {b:.6f} , mse = {mse:.6f}')
    return w,b,mse
```

```python
w,b,mse = linear_gradient_descent(0.01, 10)
```

```python
y_hat = w*np.array([0,1]) + b 
plt.scatter(X, y, s=5, color='r')
plt.plot([0,1], y_hat, color='k')
plt.title(f'{n} point of data with 10 epochs and mse = {mse:.4f}')
plt.xlabel('x')
plt.ylabel('y');
```

```python
w,b,mse = linear_gradient_descent(0.01, 100)
y_hat = w*np.array([0,1]) + b 
plt.scatter(X, y, s=5, color='r')
plt.plot([0,1], y_hat, color='k')
plt.title(f'{n} point of data with 100 epochs and mse = {mse:.4f}')
plt.xlabel('x')
plt.ylabel('y');
```
